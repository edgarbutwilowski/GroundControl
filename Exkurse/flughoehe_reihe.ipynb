# Flughöhe von Modellraketen berechnen

**1. Einführung**

In diesem Workshop werden wir die Grundlagen der Berechnung der Flughöhe von Modellraketen kennenlernen. Wir werden ein vereinfachtes Modell verwenden, das Gravitation, Gewicht und Raketenantrieb berücksichtigt, aber den Luftwiderstand vernachlässigt.

**2. Benötigte Bibliotheken**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**3. Modellraketenmotoren**

Wir werden die folgenden Modellraketenmotoren in unseren Berechnungen verwenden:

* **A8-3**: Kleiner Motor für leichte Raketen
* **B4-4**: Mittelstarker Motor für etwas größere Raketen
* **C6-5**: Leistungsstärkerer Motor für größere und schwerere Raketen
* **D12-0**: Leistungsstarker Motor für sehr große Raketen.

Hier sind die ungefähren Spezifikationen dieser Motoren:

* **A8-3:**
    * Schub: 8 N
    * Brenndauer: 0,5 s
* **B4-4:**
    * Schub: 4 N
    * Brenndauer: 1 s
* **C6-5:**
    * Schub: 6 N
    * Brenndauer: 1,7 s
* **D12-0:**
    * Schub: 12 N
    * Brenndauer: 2,5 s

**4. Berechnung der Flughöhe**

```python
def berechne_flughoehe(masse, schubkraft, brenndauer):
    g = 9.81  # Erdbeschleunigung
    nettokraft = schubkraft - (masse * g)
    beschleunigung = nettokraft / masse
    geschwindigkeit_brenn = beschleunigung * brenndauer
    hoehe_brenn = 0.5 * beschleunigung * brenndauer**2
    hoehe_frei = geschwindigkeit_brenn**2 / (2 * g)
    gesamthoehe = hoehe_brenn + hoehe_frei
    return gesamthoehe
```

**5. Beispielberechnungen**

```python
# Raketendaten
masse_rakete = 0.1  # Masse der Rakete in kg

# Berechnungen für verschiedene Motoren
motor_a8_3 = berechne_flughoehe(masse_rakete, 8, 0.5)
motor_b4_4 = berechne_flughoehe(masse_rakete, 4, 1)
motor_c6_5 = berechne_flughoehe(masse_rakete, 6, 1.7)
motor_d12_0 = berechne_flughoehe(masse_rakete, 12, 2.5)

# Ausgabe
print(f"Flughöhe mit A8-3: {motor_a8_3:.2f} m")
print(f"Flughöhe mit B4-4: {motor_b4_4:.2f} m")
print(f"Flughöhe mit C6-5: {motor_c6_5:.2f} m")
print(f"Flughöhe mit D12-0: {motor_d12_0:.2f} m")
```

**6. Variation der Raketenmasse**

Wir können die Auswirkungen der Raketenmasse auf die Flughöhe untersuchen:

```python
massen = np.linspace(0.05, 0.5, 10)  # Variation der Masse
hoehen_a8_3 = [berechne_flughoehe(m, 8, 0.5) for m in massen]
hoehen_b4_4 = [berechne_flughoehe(m, 4, 1) for m in massen]
hoehen_c6_5 = [berechne_flughoehe(m, 6, 1.7) for m in massen]
hoehen_d12_0 = [berechne_flughoehe(m, 12, 2.5) for m in massen]

# Diagramm
plt.plot(massen, hoehen_a8_3, label="A8-3")
plt.plot(massen, hoehen_b4_4, label="B4-4")
plt.plot(massen, hoehen_c6_5, label="C6-5")
plt.plot(massen, hoehen_d12_0, label="D12-0")
plt.xlabel("Raketenmasse (kg)")
plt.ylabel("Flughöhe (m)")
plt.legend()
plt.show()
```
